```python
import sys
#sys.argv[1]
with open('cnn_0005.gold_conll','r',encoding='utf-8') as conll:
    conll=list(conll)
    cpt=0
    heyho=[]
    for e in conll:
        if e!='\n' and e[0]!='#':
            cpt+=1
            b=e.split()
            heyho.append(b)
    print(conll[-3], cpt)
    different=[]
    for e in heyho:
        if e[-1]!='-':
            hm=str(e[-1]).replace('(','').replace(')','').strip()
            different.append(hm)
    print(len(set(different)))
            
                
```
